#### Plot the Amplitude Spectral Density (ASD)
Plotting these data in the Fourier domain gives us an idea of the frequency content of the data. A way to visualize the frequency content of the data is to plot the amplitude spectral density, ASD.

The ASDs are the square root of the power spectral densities (PSDs), which are averages of the square of the fast fourier transforms (FFTs) of the data.

They are an estimate of the "strain-equivalent noise" of the detectors versus frequency, which limit the ability of the detectors to identify GW signals.

They are in units of mv/rt(Hz). So, if you want to know the root-mean-square (rms) channel noise in a frequency band, integrate (sum) the squares of the ASD over that band, then take the square-root.

```python
# sampling rate:
fs = 4096
# number of sample for the fast fourier transform:
NFFT = 1*fs
fmin = 10
fmax = 2000
Pxx_H1, freqs = mlab.psd(df['LA-RA'], Fs = fs, NFFT = NFFT) # psd Compute the power spectral density ##https://matplotlib.org/api/mlab_api.html#matplotlib.mlab.psd
Pxx_L1, freqs = mlab.psd(df['LL-RA'], Fs = fs, NFFT = NFFT)

# We will use interpolations of the ASDs computed above for whitening:
psd_H1 = interp1d(freqs, Pxx_H1)
psd_L1 = interp1d(freqs, Pxx_L1)

# plot the ASDs:
plt.figure()
plt.loglog(freqs, np.sqrt(Pxx_H1),'r',label='LA-RA')
plt.loglog(freqs, np.sqrt(Pxx_L1),'g',label='LL-RA')
plt.axis([fmin, fmax, 1e-1, 1e-5])
plt.grid('on')
plt.ylabel('ASD (mv/rtHz)')
plt.xlabel('Freq (Hz)')
plt.legend(loc='upper center')
plt.title('ECG Data')
```

```python
# plot +- 5 seconds around the event:
tevent = df['Timestamp'][200]     #'2015-06-25 15:39:24.745000' 
deltat = datetime.timedelta(seconds= 10)                     # 10 seconds around the event
# index into the strain time series for this time interval:
indxt = np.where((df['Timestamp'] >= tevent-deltat) & (df['Timestamp'] < tevent+deltat))
#fs # the sampling rate
fs = 500

# pick a shorter FTT time interval, like 1/8 of a second:
NFFT = fs/8 # the length of the windowing segment
# and with a lot of overlap, to resolve short-time features:
NOVL = NFFT*15/16
# and choose a window that minimizes "spectral leakage" 
# (https://en.wikipedia.org/wiki/Spectral_leakage)
window = np.blackman(NFFT)

# the right colormap is all-important! See:
# http://matplotlib.org/examples/color/colormaps_reference.html
# viridis seems to be the best for our purposes, but it's new; if you don't have it, you can settle for ocean.
spec_cmap='viridis'
#spec_cmap='ocean'
#Pxx, freqs, bins, im = plt.specgram(df.loc[indxt]['LA-RA'], NFFT=NFFT,   Fs=Fs,noverlap=100, cmap=plt.cm.gist_heat)
plt.show() 

# Plot the LA-RA spectrogram:
plt.figure()
Pxx, freqs, bins,im = plt.specgram(df.loc[indxt]['LA-RA'], NFFT=200, Fs=fs,noverlap=100)
# spec_H1, freqs, bins, im = plt.specgram(df.loc[indxt]['LA-RA'][3:], NFFT=NFFT, Fs=fs, window=window, 
#                                         noverlap=NOVL, cmap=spec_cmap, xextent=[-deltat,deltat])
plt.xlabel('time (s) since '+str(tevent))
plt.ylabel('Frequency (Hz)')
plt.colorbar()
plt.title('LA-RA data')

# Plot the LA-RA spectrogram:
plt.figure()
Pxx, freqs, bins,im = plt.specgram(df.loc[indxt]['LL-RA'], NFFT=200, Fs=fs,noverlap=100)
# spec_H1, freqs, bins, im = plt.specgram(df.loc[indxt]['LA-RA'][3:], NFFT=NFFT, Fs=fs, window=window, 
#                                         noverlap=NOVL, cmap=spec_cmap, xextent=[-deltat,deltat])
plt.xlabel('time (s) since '+str(tevent))
plt.ylabel('Frequency (Hz)')
plt.colorbar()
plt.title('LA-RA data')

```
