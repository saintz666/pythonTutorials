```python
import numpy as np
import pandas as pd
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = 'https://en.wikipedia.org/wiki/Regression_analysis'
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#tips vs total bill
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws ={'marker': 'o','color':'indianred'},
          line_kws = {'linewidth': 1,'color': 'blue'})
```

```python
#multiple higher orders for quadrartic fits specify a order fit and not only linear
sns.lmplot('total_bill','tip',tips,order = 4,
          scatter_kws ={'marker': 'o','color':'indianred'},
          line_kws = {'linewidth': 1,'color': 'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg = False)
# remove the regression
          
```

```python
# discrete variables such as percentage of tip
tips.head()
```

```python
tips['tip_pect'] = 100 * (tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
#littler easier to read.. use Jitter
url ='https://en.wikipedia.org/wiki/Jitter'
```

```python
sns.lmplot('size','tip_pect',tips,x_jitter = 0.1)
# points do not overlap each other
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator = np.mean)
# estimate of tendency of each bin
```

```python
#use the hue facet
# two scatter plots and regression plots 
sns.lmplot('total_bill','tip_pect',tips,hue = 'sex',markers = 
          ['x','o'])
```

```python
#does day make a difference
sns.lmplot('total_bill','tip_pect',tips,hue ='day')
```

```python
#local regression model fitting
url = 'https://en.wikipedia.org/wiki/Local_regression'
```

```python
sns.lmplot('total_bill','tip_pect',tips,lowess= True,line_kws={
        'color': 'black'
    })
```

```python
#reg plot is for regression plot : it is lower level plot over which lm plot is built on
sns.regplot('total_bill','tip_pect',tips)
```

```python
fig,(axis1,axis2) = plt.subplots(1,2,sharey = True)
sns.regplot('total_bill','tip_pect',tips,ax = axis1)
sns.violinplot(tips['size'],tips['tip_pect'],palette='Reds_r',ax = axis2)
```
