```python
import numpy as np
import pandas as pd
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = randn(100)
```

```python
#displot allow combine plot
sns.distplot(dataset,bins = 25)
```

```python
#choose rugplot and others etc.
sns.distplot(dataset,bins = 25,rug = True,hist=False)
```

```python
#control the plots pass dictionary
sns.distplot(dataset,bins = 25,
             #edit kdeplot
             kde_kws = {'color':'indianred','label':'KDE PlOT'},
             #edit hist plot
             hist_kws = {'color':'green','label':'HIST'})
```

```python
#use pandas data objects
from pandas import Series
ser1 = Series(dataset,name ='My_Data')
ser1
```

```python
sns.distplot(ser1,bins = 25)
```
