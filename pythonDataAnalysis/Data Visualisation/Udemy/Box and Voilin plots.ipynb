```python
import numpy as np
import pandas as pd
from numpy.random import randn
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = 'https://en.wikipedia.org/wiki/Box_plot'
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
#have to use data = for two to show
url = 'https://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.boxplot.html'
sns.boxplot(data = [data1,data2])
```

```python
#to include the ouliers
sns.boxplot(data =[data1,data2],whis = np.inf)
```

```python
sns.boxplot(data1,orient = 'h') # for verticle axis
```

```python
# violin or viola plots .. box plot with kde plot
# box plot doesnot give a full picture
data1 = stats.norm(0,5).rvs(100)

# two gama distributions and concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50) -1,
                       -1 * stats.gamma(5).rvs(50)])
#box plot data1 and data 2
sns.boxplot(data = [data1,data2],whis = np.inf)
```

```python
# to get a full picture 
sns.violinplot(data =[data1,data2],orient = 'v')
```

```python
sns.violinplot(data2,bw=0.01)# enteer bandwidth
```

```python

sns.violinplot(data1)
```

```python
#same as rug plot, individual lines are data points
sns.violinplot(data1,inner ='stick' )
```
