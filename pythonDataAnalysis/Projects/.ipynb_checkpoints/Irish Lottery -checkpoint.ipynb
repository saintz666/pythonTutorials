```python
#import the library used to query a website
import requests


```

```python
#specify the url
lottery = "https://www.irishlottery.com/archive-2017"
```

```python
#Query the website and return the html to the variable 'page'
page = requests.get(lottery)
```

```python
page
```

```python
page.status_code
```

A status_code of 200 means that the page downloaded successfully. We wonâ€™t fully dive into status codes here, but a status code starting with a 2 generally indicates success, and a code starting with a 4 or a 5 indicates an error.

```python
#print out the HTML content of the page using the content property
page.content
```

```python
#import the Beautiful soup functions to parse the data returned from the website
from bs4 import BeautifulSoup
```

```python
#Parse the html in the 'page' variable, and store it in Beautiful Soup format
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#print out the HTML content of the page, formatted nicely
print(soup.prettify())
```

```python
print(page)
```

```python
ball_class = soup.find_all('td', class_='ball')
ball = [pt.get_text() for pt in ball_class]
ball
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]
```

```python
len(ball)
```

```python
import pprint
pprint.pprint(list(chunks(ball, 5)))
```
