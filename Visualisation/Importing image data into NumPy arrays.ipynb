```python
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline

# load already prepared ndarray from scipy
lena = scipy.misc.ascent()

# set the default colormap to gray
plt.gray()

plt.imshow(lena)
plt.colorbar()
plt.show()
```

This should open a new window with a figure displaying ascent's image in gray tones and axes. The color bar shows a range of values in the figure; here it shows 0—black to 255—white.
